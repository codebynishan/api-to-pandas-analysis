```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import requests
```

```python
data_url ="https://sdp-prem-prod.premier-league-prod.pulselive.com/api/v1/competitions/8/seasons/2025/teams?_limit=20"
r = requests.get(data_url)
if r.status_code==200:
    data = r.json()["data"]
    
else:
    print("error in fetching data")
    
print(data)
```

```python
df=pd.DataFrame(data)
df
```

```python
df.isnull().any()
```

```python
df_stadium = pd.json_normalize(df["stadium"]).add_prefix("stadium.")

```

```python
df_flat = pd.concat([df.drop(columns=["stadium"]), df_stadium], axis=1)
df_flat
```
